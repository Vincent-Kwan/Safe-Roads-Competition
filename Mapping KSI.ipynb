```python
def plotDot(point):
    folium.Marker(location=[point['LATITUDE'], point['LONGITUDE']],
                        icon=folium.Icon(icon = 'glyphicon glyphicon-remove')
   ).add_to(map)
```

```python
def plotCircle(point):
    folium.CircleMarker(location=[point['LATITUDE'], point['LONGITUDE']],
                        radius = point['COUNT'],
                        color = 'purple',
                        opacity = 0.4,
                        fill_opacity = 0.3,
                        fill = True
   ).add_to(map)
```
